# Code for System of Differnital Equations
#### Lecture at class

$\frac{dx} {dt} =3x - 4y$

$\frac{dy} {dt} =4x - 7y$

Inital Values

$x(0) =y(0) = 1 $

The Soln of equations

$ x =\frac{2}{3} e^t +\frac{1}{3} e^{-5t}$


$ y =\frac{1}{3}e^t +\frac{2}{3} e^{-5t} $

$x(0) = y(0) = 1 $

```python
import numpy as np
import matplotlib.pyplot as plt
def dxdt(x, t):
    return 1/3*np.exp(-5*t) + 2/3*np.exp(t)

def dydt(y, t):
    return 1/3*np.exp(t) + 2/3*np.exp(-5*t)

# Initial conditions
x0 = 1
y0 = 1

# Time span
t_span = np.linspace(0, 5, 1000)

# Solve the system of differential equations
from scipy.integrate import odeint
x_sol = odeint(dxdt, x0, t_span)
y_sol = odeint(dydt, y0, t_span)

# Plot the solutions
plt.plot(t_span, x_sol, label='x(t)')
plt.plot(t_span, y_sol, label='y(t)')
plt.xlabel('t')
plt.ylabel('x(t), y(t)')
plt.title('Solutions to the System of Differential Equations')
plt.legend()
plt.show()
```
